```python
import pandas as pd
import csv
```

```python
data = [
    ("march 6", 310),
    ("march 7", 340),
    ("march 8", 380),
    ("march 9", 302),
    ("march 10", 297),
    ("march 11", 323)
]
```

```python
df = pd.DataFrame(data, columns=["date", "value"])
df.to_csv("data.csv", index=False, header=None)
# Read the CSV file and below code will remove header row names
# header = df.iloc[0].tolist()
# df=df[1:]
# df.columns=header
print("created data.csv file successfully")
print(df)
```

```python
stock_prices = []
with open("data.csv", "r") as file:
    for line in file:
        tokens = line.split(",")
        day=tokens[0]
        price=int(tokens[1])
        stock_prices.append([day, price])
```

in above code we read file line by line and made 2d list which contains days and prices.

```python
print(stock_prices)
```

```python
# Actually we use below because we store stock prices in list. that's why we need to loop through it to find price on march 11
for elemnt in stock_prices:
    if elemnt[0]=="march 11":
        print("price on march 11 is:", elemnt[1])
```

```python
# if we use dictionary stock prices will be like below the code is easier.
stock_prices = {}
with open("data.csv", "r") as file:
    for line in file:
        day, price = line.split(",")
        price = int(price)
        stock_prices[day] = price
```
